```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.30,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
train_scale = scaler.fit_transform(x_train)
```

```python
test_scale = scaler.transform(x_test)
```

```python
scaled_train_df = pd.DataFrame(train_scale, columns = x_train.columns,index = x_train.index)
```

```python
scaled_test_df = pd.DataFrame(test_scale, columns = x_test.columns, index = x_test.index)
```

```python
scaled_train_df
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_nb = GaussianNB()
```

model_rf = RandomForestClassifier()

```python
model_rf = RandomForestClassifier()
```

```python
model_nb.fit(scaled_train_df, y_train)
```

```python
model_rf.fit(scaled_train_df, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
y_pred_nb = model_nb.predict(scaled_test_df)
```

```python
print(accuracy_score(y_pred_nb,y_test))
```

```python
y_pred_rf = model_rf.predict(scaled_test_df)
```

```python
print(accuracy_score(y_pred_rf,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'var_smoothing':[0.1,0.01,0.5,0.005,0.001,1e-6,1e-8,1e-10,1e-11]}
```

```python
grid = GridSearchCV(estimator=model_nb,param_grid = param_grid, cv = 5 , verbose = 3)
```

```python
grid.fit(scaled_train_df,y_train)
```

```python
grid.best_params_
```

```python
model_nb_new = GaussianNB(var_smoothing=0.5)
model_nb_new.fit(scaled_train_df, y_train)
y_pred_new = model_nb_new.predict(scaled_test_df)
print(accuracy_score(y_test, y_pred_new))
```
